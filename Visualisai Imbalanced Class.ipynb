```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('MICE2024.csv')
```

```python
#menghitung distribusi categori
count = df['categori'].value_counts()
```

```python
#visualisasi distribusi categori dengan diagram lingkaran
plt.figure(figsize=(6,6))
plt.pie(count, labels=count.index, autopct='%1.1f%%', colors=['#66b3ff','#99ff99','#ff6666'], startangle=90)
plt.show()
```
